```python
import os
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
from fastai.vision.all import *
```

```python
# Function to plot mel spectrograms and save as image
def plot_and_save_mel_spectrogram(file_path, output_folder):
    # Load audio file
    y, sr = librosa.load(file_path, sr=None)

    # Plot mel spectrogram
    plt.figure(figsize=(10, 4))
    plt.specgram(y, NFFT=2048, Fs=sr, Fc=0, noverlap=1024, cmap='viridis', aspect='auto')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')

    # Save mel spectrogram as an image file with the same name as the original .wav file
    output_file = os.path.join(output_folder, os.path.basename(file_path)[:-4] + '.png')
    plt.savefig(output_file, bbox_inches='tight')
    plt.close()
```

```python
# Specify the path to your folder containing .wav files
folder_path = 'recordings'

# Specify the path to the output folder for saving mel spectrograms
output_folder = 'output\spectograms'

# Get a list of all .wav files in the folder
wav_files = [file for file in os.listdir(folder_path) if file.endswith('.wav')]

# Iterate through each .wav file in the folder
for wav_file in wav_files:
    # Construct the full path for the .wav file
    file_path = os.path.join(folder_path, wav_file)

    # Plot and save mel spectrogram for the current .wav file
    plot_and_save_mel_spectrogram(file_path, output_folder)
```
